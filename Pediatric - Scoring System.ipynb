```python
#healthDF.loc[healthDF['46_Smoker_never smoked'] == 1, 'smoking_score'] += 100

def update_smoking_score(row):
    if row['46_Smoker_never smoked'] == 1 or '45_Smoking_home_no' == 1:
        return row['smoking_score'] + 100
    else: 
        return row['smoking_score']
    return row['smoking_score']
        
# Apply the function to the DataFrame
healthDF['smoking_score'] = healthDF.apply(update_smoking_score, axis=1)
```

## Calculate Sleeping Score

```python
def update_sleeping_score(row):
    if row['Age'] <= 12:
        if row['42_Sleep_per_day_>=9'] == 1:
            return row['sleeping_score'] + 100
        elif row['42_Sleep_per_day_7-8 hours'] == 1:
            return row['sleeping_score'] + 75
        elif row['42_Sleep_per_day_6'] == 1:
            return row['sleeping_score'] + 50
        elif row['42_Sleep_per_day_<=5'] == 1:
            return row['sleeping_score'] + 25
    else:
        if row['42_Sleep_per_day_>=9'] == 1:
            return row['sleeping_score'] + 100
        elif row['42_Sleep_per_day_7-8 hours'] == 1:
            return row['sleeping_score'] + 100
        elif row['42_Sleep_per_day_6'] == 1:
            return row['sleeping_score'] + (100/3*2)
        elif row['42_Sleep_per_day_<=5'] == 1:
            return row['sleeping_score'] + (100/3)
    return row['sleeping_score']

# Apply the function to the DataFrame
healthDF['sleeping_score'] = healthDF.apply(update_sleeping_score, axis=1)
```

## Calculate Physical Activity Score

```python
def update_physical_score1(row):
    if row['36_Exercise_per_wk_< 1 hour per week'] == 1:
        return row['physical_score'] + (50/5)
    elif row['36_Exercise_per_wk_1-2 hours per week'] == 1:
        return row['physical_score'] + (50/5*2)
    elif row['36_Exercise_per_wk_3-4 hours per week'] == 1:
        return row['physical_score'] + (50/5*3)
    elif row['36_Exercise_per_wk_5-6 hours per week'] == 1:
        return row['physical_score'] + (50/5*4)
    elif '36_Exercise_per_wk_>=7 hours per week' in healthDF.columns and row['36_Exercise_per_wk_>=7 hours per week'] == 1:
        return row['physical_score'] + 50
    return row['physical_score']
    
def update_physical_score2(row):
    if row['40_Week_exercise_hours_>=2 hours per day'] == 1:
        return row['physical_score'] + 5
    elif row['40_Week_exercise_hours_3-4 hours per day'] == 1:
        return row['physical_score'] + 7.5
    elif row['40_Week_exercise_hours_5-6 hours per day'] == 1:
        return row['physical_score'] + 10
    elif row['40_Week_exercise_hours_>=7 hours per day'] == 1:
        return row['physical_score'] + 12.5
    #elif row['40_Week_exercise_hours_none'] == 1:
    #    return row['physical_score'] + 2.5
    return row['physical_score']

def update_physical_score3(row):
    if row['41_Weekend_exercise_hours_>=2 hours per day'] == 1:
        return row['physical_score'] + 5
    elif row['41_Weekend_exercise_hours_3-4 hours per day'] == 1:
        return row['physical_score'] + 7.5
    elif row['41_Weekend_exercise_hours_5-6 hours per day'] == 1:
        return row['physical_score'] + 10
    elif row['41_Weekend_exercise_hours_>=7 hours per day'] == 1:
        return row['physical_score'] + 12.5
    #elif row['41_Weekend_exercise_hours_none'] == 1:
    #    return row['physical_score'] + 2.5
    return row['physical_score']
        
# Apply the function to the DataFrame
healthDF['physical_score'] = healthDF.apply(update_physical_score1, axis=1) + healthDF.apply(update_physical_score2, axis=1) + healthDF.apply(update_physical_score3, axis=1)
```
